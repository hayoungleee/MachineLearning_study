```python
from sklearn import datasets
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from IPython.display import display
```

```python
iris = datasets.load_iris()
print(type(iris))
```

```python
print('data shape :',iris.data.shape)
print('iris target :',iris.target_names)
print('iris features :',iris.feature_names)

X = iris.data # 데이터(특성,변수)
print('type x:', type(X)) # 타입을 확인하는 과정을 중요하다 array인지 list 인지
print(X[:5])
y = iris.target # 분류클래스 (레이블 종속 y)
print('type y: ', type(y))
print(y[:5])
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
scaler = StandardScaler()
scaler.fit(x_train,y_train)
x_train_scale  = scaler.transform(x_train)
x_test_scale = scaler.transform(x_test)
print(x_train_scale)
print(x_test_scale)
# scale을 해주는 이유 : overfitting 방지, 연산속도 상승 등등.
```

```python
# 머신러닝 모델 선택.
gnb = GaussianNB()
gnb.fit(x_train_scale,y_train) # 모델 학습
y_pred = gnb.predict(x_test_scale) # 예측
display(confusion_matrix(y_test,y_pred)) # 분류모델의 성능평가. 오차행렬
# y(specise 의 클래스가 3가지 이므로 3x3 이 나온것.)
print(classification_report(y_test,y_pred))
```

```python
# 와인 데이터 셋을 이용하여 나이브즈 
wine = datasets.load_wine()
print("Features: ", wine.feature_names)
print("Labels: ", wine.target_names)
```

```python
X = wine.data
y = wine.target
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
```

```python
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_pred = gnb.predict(X_test)
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

```python
# cross validation 를 이용한 성능평가 (정확도) 96프로 정도를 설명할 있다.
# x변수를 이런~이런걸 사용해서 y를설명하는데 설명력 
scores = cross_val_score(gnb,X,y,cv=5,scoring='accuracy') 
print(scores)
print(scores.mean())
```
