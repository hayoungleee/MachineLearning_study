```python
import numpy as np
import pandas as pd
```

```python
data = { 'x1': [13,np.nan,17,20,22,21,11,56,999,64],
        'x2': [9,555,17,11,np.nan,10,17,777,22,34],
       'y': [20,np.nan,30,27,np.nan,32,17,20,22,899]}
# 극단치와 결측치가 포함된 데이터프레임 df 를생성.
df = pd.DataFrame(data)
df
```

```python
# na값을 0으로 fill 
df_0 = df.fillna(0)
df_0
```

```python
# null 값을 채워주는 메소드 1) backfill or bfill back값으로 채운다.
df1 = df.fillna(method='bfill') 
df2 = df.fillna(method='backfill')
print(df1)
print("-"*40)
print(df2)
```

```python
# where 절을 이용하는 방법과 mean 을 이용한 방법.
print(df)
print("-"*40)
a = df.fillna(df.mean())
b = df.where(pd.notnull(df),df.mean(),axis = 1) 

print(a)
print("-"*40)
print(b)
```

```python
print(df)
print("-"*40)
a = df.fillna(df.median())
b = df.where(pd.notnull(df),df.median(),axis = 1) # true인 컬럼(1)값에 대하여 median값으로 
print(a)
print("-"*40)
print(b)
```

```python
# 판다스의 디폴트는 0 즉, 열을 기준으로 
# 판다스 where 과 넘파이 where 을  비교해보자.
print(df)
print("-"*40)
row_mean_data = df.mean(axis=1) # 행의 평균
print(row_mean_data)
print("-"*40)
col_mean_data = df.mean() # 열의 평균
print(col_mean_data)
```
